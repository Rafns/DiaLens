```python
import joblib
import pandas as pd

# 1. MUAT MODEL & AMBIL DAFTAR FITUR
try:
    model_nonlab = joblib.load('model_prediksi_diabetes_nonlabv1.joblib')
    model_gabungan = joblib.load('model_prediksi_diabetes_gabunganv1.joblib')
    
    # Ambil fitur dan konversi ke str biasa (hindari np.str_)
    FEATURE_NON_LAB = [str(f) for f in model_nonlab.feature_names_in_]
    FEATURE_GABUNGAN = [str(f) for f in model_gabungan.feature_names_in_]
    
    print("âœ… Model berhasil dimuat.")
    print("Fitur Non-Lab:", FEATURE_NON_LAB)
    print("Fitur Gabungan:", FEATURE_GABUNGAN)
    
except Exception as e:
    print(f"âŒ Gagal memuat model: {e}")
    exit()

# 2. FUNGSI PREDIKSI
def predict_nonlab(data_dict):
    """Prediksi menggunakan model non-lab (9 fitur)."""
    try:
        df = pd.DataFrame([data_dict], columns=FEATURE_NON_LAB)
        prob = model_nonlab.predict_proba(df)[0, 1]
        return prob
    except Exception as e:
        print(f"âŒ Error prediksi non-lab: {e}")
        return 0.5

def predict_gabungan(data_dict):
    """Prediksi menggunakan model gabungan (11 fitur)."""
    try:
        df = pd.DataFrame([data_dict], columns=FEATURE_GABUNGAN)
        prob = model_gabungan.predict_proba(df)[0, 1]
        return prob
    except Exception as e:
        print(f"âŒ Error prediksi gabungan: {e}")
        return 0.5

# 3. DATA UJI â€” SESUAIKAN NILAI DI SINI
# Contoh 1: Data risiko tinggi
data_tinggi = {
    'Age': 33.0,
    'DietQuality': 5,                
    'Smoking': 1,                    
    'FamilyHistoryDiabetes': 1,      
    'Hypertension': 1,               
    'FrequentUrination': 1,          
    'ExcessiveThirst': 1,            
    'UnexplainedWeightLoss': 1,      
    'HealthLiteracy': 5,             
    'FastingBloodSugar': 120.0,      
    'HbA1c': 4                   
}

# Contoh 2: Data risiko rendah
data_rendah = {
    'Age': 20.0,
    'DietQuality': 0,
    'Smoking': 1,
    'FamilyHistoryDiabetes': 0,
    'Hypertension': 0,
    'FrequentUrination': 0,
    'ExcessiveThirst': 0,
    'UnexplainedWeightLoss': 0,
    'HealthLiteracy': 0,
    'FastingBloodSugar': 50.0,
    'HbA1c': .0
}

# Pilih data uji di sini
data_uji = data_tinggi  

# 4. JALANKAN PREDIKSI
print("\n" + "="*60)
print("ðŸ“Š UJI PREDIKSI MANUAL")
print("="*60)

# Prediksi Non-Lab (hanya 9 fitur)
prob_nl = predict_nonlab(data_uji)
risk_nl = "Tinggi" if prob_nl >= 0.7 else ("Sedang" if prob_nl >= 0.5 else "Rendah")
print(f"\nðŸ” Model Non-Lab (9 Fitur):")
print(f"   Probabilitas: {prob_nl:.4f} ({prob_nl:.2%})")
print(f"   Risiko: {risk_nl}")

# Prediksi Gabungan (11 fitur)
prob_gab = predict_gabungan(data_uji)
risk_gab = "Tinggi" if prob_gab >= 0.6 else ("Sedang" if prob_gab >= 0.5 else "Rendah")
print(f"\nðŸ§ª Model Gabungan (11 Fitur):")
print(f"   Probabilitas: {prob_gab:.4f} ({prob_gab:.2%})")
print(f"   Risiko: {risk_gab}")

print("\n" + "="*60)
```
